# Statistische Grundlagen

Was ist Statistik?

Statistik „ist die Lehre von Methoden zum Umgang mit quantitativen Informationen“. Sie ist eine Möglichkeit, "eine systematische Verbindung zwischen Erfahrung und Theorie herzustellen“. Unter Statistik versteht man die Zusammenfassung bestimmter Methoden zur Analyse empirischer Daten. 

Wir unterscheiden zwischen deskriptiver Statistik und Inferenststatistik:

Deskriptive Statistik
 - Beschreibende und zusammenfassend Statistiken 
 - Anteile

Inferenzstatistik
 - Stichprobendaten werden verwendet um Aussagen über größere      
 - Zusammenhänge zu analysieren.

**Datentypen**

Datentypen bestimmen maßgeblich, welche Methoden und Visualisierungen wir für die Datenanalyse verwenden.

Numerische Daten (Quantitativ)
 - Kontinuierliche Variablen (Teilbar)
  - Zeit 
  - Geschwindigkeit 

 - Diskrete Variablen (Zählbar)
  - Anzahl der Haustiere 
  - Anzahl der versendeten Packete

Kategoriale Daten (Qualitativ)
  - Nominal (Ungeordnet)
   - Binärdaten
   - Aufenthaltsland

  - Ordinal (Geordnet)
   - Fragebogenzustimmung






```python
# Numerische Daten 

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from vega_datasets import data
import statistics 

cars = data.cars()
np.mean(cars['Horsepower'])

plt.scatter(x='Horsepower', y='Acceleration', data= cars, alpha=0.5)
plt.show()
```

```python
# Kategoriale Daten

cars['Origin'].value_counts()
sns.countplot(cars['Origin'])
plt.show()
```
