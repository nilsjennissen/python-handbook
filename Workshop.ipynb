**7. Daten anpassen**

```python
cars1 = cars.sort_values("Horsepower", ascending=False)
cars1.head()
```

```python
cars2 = cars.sort_values(["Horsepower", "Weight_in_lbs"], ascending=[False, True])
cars2.head()
```

```python
is_German = cars1["Origin"] == "Germany"
is_not_8Cylinder = cars1["Cylinders"] != 8
cars3 = cars2[is_German & is_not_8Cylinder]
cars3
```

**5. Visualisierungen**

Suchen


```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Horsepower', y='Origin', data=cars, hue='Cylinders')
```

```python
sns.regplot(x="Horsepower", y="Cylinders", data=cars)
```

```python
plt.figure(figsize=(10,10))
sns.set_theme(style="whitegrid")


cmap = sns.cubehelix_palette(rot=-.2, as_cmap=True)
g = sns.relplot(
    data=cars,
    x="Acceleration", y="Horsepower",
    hue="Origin", size="Cylinders"
)
g.set(xscale="log", yscale="log")
g.ax.xaxis.grid(True, "minor", linewidth=.25)
g.ax.yaxis.grid(True, "minor", linewidth=.25)
g.despine(left=True, bottom=True)
```
